## 1. Survived

##Survived Value Counts

**conclusion**
- The number of passengers in the Titanic training dataset who did not survive was 549 (61.6%)
- And those who survived was 342 (38.4%).

```python
df['Survived'].value_counts()
```

##Survived Bar plot

```python
Survived=df['Survived'].value_counts()
plt.bar(Survived.index,Survived.values, color="green")
plt.title('Bar graph Of Survived People')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0,1],[0,1])
plt.show()
```

##To See Percentage% People Die, We use Pie Chart




```python

survived=df['Survived'].value_counts()
plt.pie(survived,labels=[0,1],autopct='%1.1f%%')
plt.title('Pie Chart Of Survived People')
plt.show()
```

##2. Pclass
